# Introduction to `geochem`

`geochem` is a Python package for exploring, preprocessing, plotting, and analyzing geochemical data. It is designed for notebook-based work with volcanic and igneous geochemical datasets.

The package works mainly with `pandas.DataFrame` objects, where rows are samples and columns are geochemical components named with conventional symbols such as `SiO2`, `FeOT`, `MgO`, `La`, `Rb`, and `Sr`. Many plotting and numerical functions also accept NumPy arrays for custom workflows.

## Download and Installation

The source code is hosted on GitHub:

<https://github.com/volcanlab/geochem>

The easiest way to install it is with `pip`:

```bash
python -m pip install git+https://github.com/volcanlab/geochem.git
```

If you use conda, first activate the environment where you want to use the package:

```bash
conda activate my_env
python -m pip install git+https://github.com/volcanlab/geochem.git
```

After installation, notebooks can import `geochem` directly.

## Import the Package

For general preprocessing and analysis tools, import the main package. For plotting notebooks, it is also useful to import `geochem.plt`.

````python
import numpy as np
import pandas as pd

import geochem as gc
import geochem.plt as gcplt
import geochem.parameters as gparam

print(f"geochem version: {gc.__version__}")
````

## Data Model

In a typical `geochem` DataFrame:

- each row is a sample;
- each geochemical component is a column;
- major oxides, trace elements, and REE use standard geochemical names;
- descriptive columns such as `Sample`, `Source`, or `Description` can be kept in the same DataFrame.

The standard names used by the package are stored in `geochem.parameters`.

````python
gc.getnames_major()
````

````python
print(f"Number of recognized geochemical columns: {len(gc.getnames_all())}")
print(f"Major oxides: {gc.getnames_major()}")
print(f"REE: {gc.getnames_REE()}")
print(f"Trace-element plotting order: {gc.getnames_trace()}")
````

## Reference Values

Some diagrams use built-in reference datasets, for example Sun and McDonough (1989) normalization values for spider diagrams and Irvine and Baragar (1971) coordinates for the AFM boundary.

````python
sm89 = gparam.getvalues_SM89("chondrite")
{k: sm89[k] for k in ["Source", "La", "Ce", "Nd", "Yb"]}
````

````python
A, F, M = gparam.getvalues_IB71()
print("First three AFM boundary coordinates:")
for row in zip(A[:3], F[:3], M[:3]):
    print(row)
````

## General Utility Functions

The `utils.py` module provides general tools used by other parts of the package. Two important examples are:

- `closure()`: rescales rows or arrays to a constant sum;
- `calculate_FeOT()`: calculates total iron as FeO from `FeO` and `Fe2O3` when `FeOT` is missing.

````python
df = pd.DataFrame({"SiO2": [50.0, 45.0], "MgO": [10.0, 5.0], "CaO": [15.0, 10.0]})
gc.closure(df, total=100)
````

````python
arr = np.array([[10.0, 40.0], [20.0, 30.0]])
gc.closure(arr, total=100)
````

````python
fe = pd.DataFrame({
    "FeO": [10.0, np.nan, 5.0],
    "Fe2O3": [5.0, 10.0, np.nan],
    "FeOT": [np.nan, 20.0, np.nan],
})

gc.calculate_FeOT(fe)
fe
````

## Preprocessing and Plotting

Most real datasets should be preprocessed before plotting. The high-level preprocessing function is:

```python
gc.preproc(df)
```

Plotting functions are available from `geochem.plt`, for example:

```python
gcplt.TAS(df)
gcplt.AFM(df)
gcplt.spider(df, elements=gcplt.getnames_trace())
gcplt.harker(df)
```
