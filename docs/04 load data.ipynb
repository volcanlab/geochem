# Loading Geochemical Data

This notebook shows how to load geochemical data into a `pandas.DataFrame`, which is the main data structure used by `geochem`.

The example files are stored in this `docs/` folder:

- `data_example.csv`
- `datos_ejemplo.xlsx`

The two files contain the same type of example geochemical table.

## Import Libraries

After installing `geochem`, import it directly. There is no need to add the package folder manually with `sys.path.append(...)`.

```python
from pathlib import Path

import pandas as pd
import geochem as gc
```

## Locate the Example Files

When a notebook is run, the current working directory can vary depending on how Jupyter was opened. Using `Path` makes the examples easier to reuse.

```python
DOCS_DIR = Path.cwd()
if not (DOCS_DIR / "data_example.csv").exists():
    DOCS_DIR = Path.cwd() / "docs"

csv_file = DOCS_DIR / "data_example.csv"
xlsx_file = DOCS_DIR / "datos_ejemplo.xlsx"

print(csv_file)
print(xlsx_file)
```

## Data Structure

`geochem` expects geochemical components to be stored in columns with standard names such as `SiO2`, `FeOT`, `MgO`, `La`, `Rb`, and `Sr`.

A table can also contain descriptive or grouping columns such as `Sample`, `Type`, `Description`, `Reference`, and `Source`. These columns are useful for filtering and plotting groups.

```python
column_names = gc.getnames_all()

print(f"Number of recognized geochemical columns: {len(column_names)}")
print(column_names)
```

## Load a CSV File

A CSV file can be loaded directly with `pandas.read_csv()`.

```python
df_csv = pd.read_csv(csv_file)
df_csv.head()
```

```python
df_csv.info()
```

The example dataset includes a `Source` column that can be used later for grouped diagrams.

```python
df_csv["Source"].unique()
```

## Load an Excel File

Excel files can be loaded with `pandas.read_excel()`. This requires the `openpyxl` package, which is commonly installed alongside notebook workflows.

```python
df_xlsx = pd.read_excel(xlsx_file)
df_xlsx.head()
```

```python
print(df_xlsx.shape)
print(df_xlsx.columns.to_list())
```

## Check Which Columns Are Geochemical Data

The table may contain both geochemical columns and metadata columns. We can compare the DataFrame columns with the names recognized by `geochem`.

```python
recognized = [col.strip() for col in df_csv.columns if col.strip() in gc.getnames_all()]
metadata = [col for col in df_csv.columns if col.strip() not in gc.getnames_all()]

print(f"Recognized geochemical columns: {len(recognized)}")
print(recognized)
print()
print("Metadata or non-standard columns:")
print(metadata)
```

Some column names in the raw file contain extra spaces, for example `SiO2 `. The preprocessing notebook shows how to clean these names before analysis.
