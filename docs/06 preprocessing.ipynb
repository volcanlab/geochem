# Preprocessing a Geochemical Dataset

Real geochemical tables often need a few cleaning steps before plotting or analysis. This notebook shows the preprocessing tools in `geochem`.

The preprocessing functions work with `pandas.DataFrame` objects and modify the DataFrame in place. This is intentional: they clean column names, convert values, and add or update geochemical columns.

## Import Libraries and Load the Dataset

After installing `geochem`, import it directly. The example data are stored in this `docs/` folder.

````python
from pathlib import Path

import numpy as np
import pandas as pd

import geochem as gc
````

````python
DOCS_DIR = Path.cwd()
if not (DOCS_DIR / "data_example.csv").exists():
    DOCS_DIR = Path.cwd() / "docs"

csv_file = DOCS_DIR / "data_example.csv"
df = pd.read_csv(csv_file)

df.head()
````

````python
df.info()
````

## Step 1: Strip Whitespace from Column Names

Some columns in the raw file have spaces at the end, for example `SiO2 `. These names will not match the standard names used by `geochem` until the spaces are removed.

````python
print("Before:")
print(df.columns.to_list())
````

````python
gc.preproc_strip_whitespace(df)
````

````python
print("After:")
print(df.columns.to_list())
````

## Step 2: Convert Geochemical Columns to Numeric Values

Geochemical data should be numeric. Some datasets include values such as `<0.5`, meaning the element was below the detection limit. `geochem` converts these values to half the detection limit, so `<0.5` becomes `0.25`.

Only recognized geochemical columns are converted. Metadata columns such as `Sample`, `Type`, and `Source` are left unchanged.

````python
print(df["Cr"].head(10))
print(df["Cr"].dtype)
````

````python
gc.preproc_make_numeric_columns(df)
````

````python
print(df["Cr"].head(10))
print(df["Cr"].dtype)
````

## Step 3: Calculate Total Iron as FeO (`FeOT`)

`calculate_FeOT()` derives total iron as FeO from `FeO` and `Fe2O3`:

```text
FeOT = FeO + 0.8998 * Fe2O3
```

When a DataFrame already has a `FeOT` column, the function only fills values that are missing or equal to zero. Existing non-zero `FeOT` values are preserved.

````python
fe_cols = [col for col in ["FeO", "Fe2O3", "FeOT"] if col in df.columns]
df[fe_cols].head(10)
````

````python
gc.preproc_calculate_feot(df)
````

````python
df[["FeO", "Fe2O3", "FeOT"]].head(10)
````

## Run All Preprocessing Steps at Once

The convenience function `preproc()` runs the three steps in order:

1. strip whitespace from column names;
2. convert recognized geochemical columns to numeric values;
3. calculate `FeOT` where needed.

````python
df2 = pd.read_csv(csv_file)
gc.preproc(df2)
````

````python
df2.info()
````
